## The Impact of Total Spend on Football League Positions ##

This analysis will look at the teams in the current season of the Premier League over the last 10 years.

The datasets have been created from total spend information provided by [transfermarkt](https://www.transfermarkt.co.uk/premier-league/fuenfjahresvergleich/wettbewerb/GB1), combined with the final league position for the given season.

Over the course of 10 years, only 7 teams have been consistently present in the top flight of English football (Arsenal, Everton, Manchester United, Manchester City, Liverpool, Tottenham and Chelsea).

The English league system promotes and relegates teams based on their league position, with the highest ranking league being the Premier League. Teams in the Championship (the next league down from the Premiership) can also be relegated, but also have the possibility of gaining promotion via end of season play-offs.

As such, to compare all the teams in the current season the final league position for the given season was taken for each of the last 10 seasons. 

Each column of the Spend dataframe is valued in millions GBP.

A total net spend is initially represented by a negative value, whereas a total net profit is represented by a positive value. This is later adjusted in the study to make comparisson easier.

The current seasons standings are accurate at the time of this report - 29/10/2020.





```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange
```

```python
# Creating dataframes
# Setting the Club column as the dataframe index

prem_spend = pd.read_excel(r"C:\Users\thazzan\Documents\Python Scripts\Comparing Premier League Spend with Performance_files\transfermarket10yr_Spend_27102020.xlsx", index_col=0)

prem_pos = pd.read_excel(r"C:\Users\thazzan\Documents\Python Scripts\Comparing Premier League Spend with Performance_files\transfermarket10yr_Position_27102020.xlsx", index_col=0)

```

```python
prem_spend
```

```python
# Total Spend by Season £m. 
#`abs()` is used to return the abolute value of the sum. As per the dataframe, 
# the dataset features negative values to represent a net spend, and positive values where a club had recorded
# a positive net transfer. `abs()` is used from here on to make the output more intuitive

prem_spend.sum().abs()
```

From summing the total spend for each season, we can see that the 20/21 season has seen the highest transfer market spend over the last 10 years (£1027.97m). 

This is interesting, epecially given that football in 2020 has been impacted by Covid-19, and spectators, a common source of income, have not been able to attend live games as normal.

```python
# highest single seasonal spend per club

prem_spend.min(axis=1).abs()
```

```python
# Total spend per club over 10 years, sorted by average spend from highest to lowest

spend_10yrs = prem_spend.sum(axis=1).sort_values(ascending=True).abs()

spend_10yrs

```

To those familiar with English football, there would be no surprise that Manchester City have the highest total spend over the last 10 years.

Manchester City are followed by Manchester United. Both of these clubs have significant financial backing, which is reflected in their purchasing power.

With the exception of Everton, the 6 highest spending clubs would also be considered the contemporary "Big 6" clubs that commonly find themselves towards the top of the league. These 6 clubs are also part of the total 7 clubs that have never been relegated from the Premier League.


```python
# reviewing league position by club

prem_pos
```

```python
# average league position over 10 years, sorted from highest average position to lowest

avg_pos = prem_pos.mean(axis=1).sort_values(ascending=True)

avg_pos
```

From the above series, we can see that over the 10 year perod reviewed, Machester City have achieved the highest average league position. 

Interstingly, Tottenham Hotspur, although appearing towards the bottom of the total spend series, are second in the average position ranking. This suggests, Tottenham Hotspur have been able to compete and outperform clubs with greater budgets consistently over the last 10 years.

This of course can be achieved by "developing" players, as opposed to purchasing them. High ticket player sales would also offset a large amount of player spend. Most notably, in 2013 Tottenham sold Gareth Bale to the Spanish team, Real Madrid, for a reported fee between 91m and 100m Euros. 

```python
# comparing Spend and Position side by side

team_table = pd.concat([spend_10yrs, avg_pos], axis=1)

team_table.columns = ["10 Year Spend", "Average Position"]
team_table.index.name = "Club"

team_table
```

By comparing the two series, there are some interesting observations. Most notably, some clubs with high levels spend over the last 10 years, have achived on average lower league positions than those that have spent less. This can be seen in the cases of Liverpool and Tottenham Hotspur.

A bar chart can be used to better visualise these relationships.

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()

team_table["10 Year Spend"].plot.bar()
ax.set_ylabel("10 Year Spend £m")

ax2 = team_table["Average Position"].plot(secondary_y=True, color='Orange', marker='o')
ax2.set_ylabel("Average Position")

plt.title("Average League Position vs Total 10 Year Spend")

ax.set_xticklabels(team_table.index, rotation=90)


plt.show()
```

From the above chart, it can be seen that the clubs with the highest spend appear to have a better average league position.

However, clubs such as Tottenham, Leicester and Liverpool appear to have higher average league positions, while maintaining a fairly low net spend over the last 10 years.

Testing for a correlation coefficient will be able to provide some more insight into the apparent relationship between spend and league position.

**Comapring The Correlations Between League Position and Spend**


```python
# Calculating a correlation between 10 Year Spend and Average Leage Position

correlation = team_table["10 Year Spend"].corr(team_table["Average Position"])

correlation
```

The correlation of 0.6 would be considered a moderate positive correlation. This falls in line with the representation seen in the above chart. Clubs that spend more money, appear to finish higher in their respective leagues, but there are enough outliers to ensure that it is not a certainty that increased spedning equals better league position.

# Conclusion #

Unsurprisingly, from the charts it can be seen that the clubs with largest spend, on average have a better league position.

The correlation of the `10 Year Spend` and `Average Position` columns indicate a moderate positive correlation. But just spending money does not equal success.

Clubs go through great effort to source, and develop talent as part of a structured recruitment strategy. The infrastructure around this, agents and a club's scouting network are also huge costs which are not reflected in the transfer fee.

Also not reflected within yearly transfer spend is how much is spent on manegerial appointments. 

In addition to expensive players, expensive managers would also be expected to deliver an improvement in league position.

Chelsea and Manchester City have all achieved Premier League titles after the appointment of high profile managers. These clubs are also notable, along with Mancheter United for having the highest net spend over the last 10 years.

Examples such as Tottenham, Leicester and Liverpool suggest that success can also be achieved by different means - player development, scouting, tactics etc. 

Liverpool won the 19/20 season, having made a net profit of £31m in the transfer window. In the 20/21 season they have spent £36m compared to Manchester City's £101m.
